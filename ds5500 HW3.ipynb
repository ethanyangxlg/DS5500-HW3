```python
district_level_performance= pd.read_csv('math-achievement-lea-sy2015-16.csv')
```

ALL_MTH00PCTPROF_1516 is a blurred metric and I will use this as my performance matric

```python
district_level_performance.groupby('ALL_MTH00PCTPROF_1516').count().reset_index()['ALL_MTH00PCTPROF_1516']
```

```python
type(district_level_performance['ALL_MTH00PCTPROF_1516'][0])
```

```python
def process_blur(data):
    for i in range(len(data)):
        if 'LT' in data[i]:
            temp = int(data[i][2:])
            data[i] = np.mean([0,temp-1]) # if it is LT (less than), take mean of 0 and that number -1 as input 
        elif 'LE' in data[i]:
            temp = int(data[i][2:])
            data[i] = np.mean([0,temp]) # if it is LE (less than or equal to), take mean of 0 and that number as input 
        elif 'GT' in data[i]:
            temp = int(data[i][2:])
            data[i] = np.mean([temp+1,100]) # if it is GT (greater than), take mean of 100 and that number +1 as input 
        elif 'GE' in data[i]:
            temp = int(data[i][2:])
            data[i] = np.mean([temp,100]) # if it is GE (less than), take mean of 100 and that number  as input 
        elif 'PS' in data[i]:
            data[i] = 'temp' # if it is suppressed to protect student privacy, wait for input
        elif '-' in data[i]:
            temp = data[i].split('-')
            data[i] = np.mean([int(temp[0]),int(temp[1])]) # if it is a range, take mean of start number and end number as input 
        else:
            data[i] = int(data[i])
    input_avg =np.mean(data[data!='temp'])
    for j in range(len(data)):
        if data[j]=='temp':
            data[j] = input_avg
    return data
```

```python
district_level_performance['ALL_MTH00PCTPROF_1516'] = process_blur(district_level_performance['ALL_MTH00PCTPROF_1516'])
```

```python
sns.distplot(district_level_performance['ALL_MTH00PCTPROF_1516'])
plt.title('Distribution of ALL_MTH00PCTPROF_1516 ')
```

# Problem 4

You are tasked with cutting 15% of the U.S. federal budget currently being spent on funding school districts. How much money is this?

```python
cut_money = 0.15*federal_fund_state['TFEDREV'].sum()
cut_money
```

First Round(National Wide 10%)

```python
school =district_level_fiscal[district_level_fiscal['TFEDREV']>0]
first_round_cut= 0.1*school['TFEDREV'].sum()
first_round_cut_list = school
first_round_cut_list['1_TFEDREV'] = 0.1*first_round_cut_list['TFEDREV']
first_round_cut_list['TFEDREV_after1'] = 0.9*first_round_cut_list['TFEDREV']
```

Second Round(Performance Focused)

```python
threshold = np.percentile(district_level_performance['ALL_MTH00PCTPROF_1516'],75)
threshold
```

```python
cut_off_list = district_level_performance[district_level_performance['ALL_MTH00PCTPROF_1516'] < threshold]
```

```python
second_round_cut_list = pd.merge(cut_off_list,first_round_cut_list, on='LEAID',how='left')
total_have =second_round_cut_list['TFEDREV_after1'].sum()
percent = 100*(cut_money-first_round_cut)/total_have
percent # cut off percent of target school
```

```python
second_round_cut_list['2_TFEDREV'] = round((percent/100)*second_round_cut_list['TFEDREV_after1'],1)
```

```python
final_cut_off = pd.merge(first_round_cut_list,second_round_cut_list,on='LEAID',how='left')
final_cut_off
```

```python
final_cut_off.fillna({'1_TFEDREV_x':0,'2_TFEDREV':0},inplace=True)
final_cut_off['Amount of Funding Cut'] = final_cut_off['1_TFEDREV_x']+final_cut_off['2_TFEDREV']
```
